```python
from ultralytics import YOLO

# Load a segmentation model (YOLOv8 small segmentation pretrained)
model = YOLO("yolov9e.pt")

# Train on your dataset
model.train(
    data="C:/Users/admin/Desktop/robo_wounds/data.yaml",  # path to your dataset yaml
    epochs=30,
    imgsz=640,
    task="detect"
)

# Evaluate on validation set
metrics = model.val()

# Run inference on test images
results = model.predict(source="C:/Users/admin/Desktop/robo_wounds/test/images")
```
